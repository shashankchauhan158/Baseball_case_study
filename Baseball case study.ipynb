Trying to find the best model for datasets



```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
imputer=imputer.fit(df)
```

```python
df=imputer.transform(df)
```

```python
df
```

```python
df=pd.DataFrame(df)
df
```

```python
y=df.iloc[:,-1]
```

```python
y.head()
```

```python
y.shape
```

```python
x=df.iloc[:,1:-1]
```

```python
x.head
```

```python
x.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.score(x_train,y_train)
```

```python
lm.coef_
```

```python
pred=lm.predict(x_test)
```

```python
print('predicted value and actual value is',pred,y_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print('Mean absolute error:',mean_absolute_error(y_test,pred))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('Root mean squared error:',np.sqrt(mean_squared_error(y_test,pred)))
```

Conclusion-In this project, we have used LinearRegression in the dataset and achieve the accuracy score. It is also used for predicting the number of wings for baseball team.
